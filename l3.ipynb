#1. Scrieți o funcție care convertește temperatura din grade celsius în fahrenheit și invers (formula: c/5 = f-32/9)

```python
#1
def fahrenheit_to_celsius(fahrenheit):
    celsius = (fahrenheit - 32) * 5 / 9
    return celsius

def celsius_to_fahrenheit(celsius):
    fahrenheit = celsius * 9 / 5 + 32
    return fahrenheit

choose = input('Enter 1 to convert Fahrenheit to Celsius.\nEnter 2 to convert from Celsius to Fahrenheit.')

match choose:
    case '1':
        fahrenheit = float(input('Enter Temperature in Fahrenheit: '))
        celsius = fahrenheit_to_celsius(fahrenheit)
        print(str(fahrenheit) + ' in F° is ' + str(round(celsius, 2)) + ' C°')
    case '2':
        celsius = float(input('Enter Temperature in Celsius: '))
        fahrenheit = celsius_to_fahrenheit(celsius)
        print(str(celsius) + ' in C° is ' + str(round(fahrenheit, 2)) + ' F°')
```

#2. Scrieți o funcție determina dacă un text este palindrom (caiac, capac, minim)

```python
#2
def palindrome(word):
    for i in range(0, len(word) // 2):
        if (word[i] != word[len(word) - i - 1]):
            return False
    return True

word = input()
palindrome(word)
```

#3. Scrieți o funcție care sa afiseze toți divizorii unui număr întreg.

```python
#3
all_divisors = []

def number_divisors(number):
    i = 1
    while i <= number :
        if (number % i == 0):
            all_divisors.append(i)
        i = i + 1

# Driver method
number = int(input())
number_divisors(number)
#print('All  divisors of ' + str(number) + ' are: ' + ', '.join(str(all_divisors)) + '.')
print(f'All divisors of {number} are: {", ".join(str(div).strip() for div in all_divisors)}.') #Understand the Problem with print() output.
```

#4. Calculati suma tuturor numerelor între 1000 si 2300 care se împart fără rest la 5 și 7.

```python
#4
floor_div_on_five = []
floor_div_on_seven = []
floor_div_on_both = []

for i in range(1000,2301):
    n = i // 5
    if (i == n * 5):
        floor_div_on_five.append(i)

for i in range(1000,2301):
    n = i // 7
    if (i == n * 7):
        floor_div_on_seven.append(i)

for i in floor_div_on_five:
    if i in floor_div_on_seven:
        floor_div_on_both.append(i)

floor_div_on_five.clear()
floor_div_on_seven.clear()

print('Sum of all values in range from 1000 to 2300 that can be floor devided by 5 and 7: '+ str(sum(floor_div_on_both)))
```

#5. Scrieți o funcție care sa calculeze numărul de litere și cifre din un text.

```python
#5
all_numbers = []
all_letters = []
all_symbols = []

#FUNCTION TO ADD ALL LETTERS, NUMBERS AND SYMBOLS TO RESPECTED LISTS
#-------------------------------------------------------------------
def add_all_characters():
    for i in range(ord('A'), ord('Z') + 1):
        all_letters.append(chr(i))

    for i in range(ord('a'), ord('z') + 1):
        all_letters.append(chr(i))

    for i in range(ord('0'), ord('9') + 1):
        all_numbers.append(chr(i))

    for i in range(ord(' '), ord('/') + 1): #rewrite symbols addition
        all_symbols.append(chr(i))

    for i in range(ord(':'), ord('@') + 1):
        all_symbols.append(chr(i))

    for i in range(ord('['), ord('`') + 1):
        all_symbols.append(chr(i))

    for i in range(ord('{'), ord('~') + 1):
        all_symbols.append(chr(i))

#FUNCTION TO COUNT ALL CHARACTERS
#-------------------------------------------------------------------
def count_letters_and_numbers(text):
    letters = 0
    numbers = 0
    symbols = 0

    for i in text:
        if i in all_letters:
            letters += 1
        elif i in all_numbers:
            numbers += 1
        else:
            symbols += 1

    return letters, numbers, symbols

#MAIN
#-------------------------------------------------------------------
add_all_characters()

text = input("Enter a text: ")

letters, numbers, symbols = count_letters_and_numbers(text)

print('Letters in text: ' + str(letters))
print('Numbers in text: ' + str(numbers))
```

#6. Scrieți un program care verifica dacă o parola introdusă de utilizator este securizata
> Lungimea minima 6 caractere

> Conține cel puțin 1 litera în intervalul [a-z]

> Conține cel puțin 1 litera în intervalul [A-Z]

> Conține cel putin 1 cifra

> Conține cel putin un caracter special [!, /, #]

> Nu contine caractere interzire [@, ‘, {, }]


```python
#6
password = input('Enter Password: ')

uppercase = 0
lowercase = 0
number = 0
a_symbols = 0
p_symbols = 0

allowed_symbols = ["!", "/", "#"]
prohibited_symbols = ["@", "'", "{", "}"]

for i in password:
    if i.isupper() == True:
        uppercase += 1
    if i.islower() == True:
        lowercase += 1
    if i.isdigit() == True:
        number += 1
    if i in allowed_symbols:
        a_symbols += 1
    if i in prohibited_symbols:
        p_symbols += 1

if len(password) >= 6 and uppercase >= 1 and lowercase >= 1 and number >= 1 and a_symbols >= 1 and p_symbols == 0:
    print(f"Password {password} is Valid")
elif p_symbols >= 1:
    print(f'There are prohibited symbols like this: {prohibited_symbols} in your Password: {password}')
elif a_symbols == 0:
    print(f'There are no allowed characters like this: {allowed_symbols} in your Password: {password}')
elif number == 0:
    print(f'There are 0 numbers in your Password: {password}')
elif uppercase == 0:
    print(f'There are 0 uppercase letters in your Password: {password}')
elif lowercase == 0:
    print(f'There are 0 lowercase letters in your Password: {password}')
elif len(password) < 6:
    print('Password lenght is less then 6 characters.')
else:
    print('Error!')
```

#7. Scrieți un program care sa elimine cuvinte duplicate dintr-o propozitie.

```python
#7 #Case sensitive
def remove_duplicates(text):
	words = text.split(" ")
	fixed_text = []
	for i in words:
		if i not in fixed_text:
			fixed_text.append(i)
	return " ".join(fixed_text)

text = input()
print(remove_duplicates(text))
```
