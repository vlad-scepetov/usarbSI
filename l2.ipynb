##6.1 Ce valoare are expresia a[int(int('3' * 2) // 11)]
> A: 'd' из списка a = ['a', 'b', 'c', 'd']

> Первым действием происходит конкатенация '3' типа String и мы получаем строку '33'.

> Вторым действием приобразуем строку '33' в тип Integer использую функцию int().

> Третьим действием делим 33 на 11 и получаем 3 и еще раз применяем функцию int().

> В итоге получаеться указатель на индекс a[3] списка a = ['a', 'b', 'c', 'd'].

##6.2 Adaugati o valoare nouă în lista a pe prima poziție.

##6.3 Ștergeți ultimele 2 valori din lista.

```python
#6.3
a = ['a', 'b', 'c', 'd']
for i in range(2):
    a.pop()
print(a)
```

##6.4 Ordonati lista a descrescator

```python
#6.4
a = ['a', 'b', 'c', 'd']
a.sort(reverse=True)
print(a)
```

##6.5 Creați o listă nouă b care sa conțină toate elementele din lista a cu excepția primelor 2 elemente.

```python
#6.5
a = ['a', 'b', 'c', 'd']
b = []

for i in range(2):
    a.pop(0)

b.append(a)
print(b)
```

#7. Scrieți un program care afișează toate elementele din un invetar si calculează suma lor.

```python
#7
inventory = {'Milk': 10, 'Bread': 11, 'Meat': 17, 'Cola': 32, 'Tea': 10, 'Cofee': 12}
items = []
quantity = []
quantity_sum = 0

for i in inventory.keys():
    items.append(i)

for i in inventory.values():
    quantity.append(i)

for i in range(len(quantity)):
    quantity_sum = quantity_sum + quantity[i]

print('Inventory Items: ' + ', '.join(items) + '.')
print('Inventory Quantity: ' + str(quantity_sum) + '.')
```
